```python
N = 100
y = np.linspace(N,1,N)
max_growth = 750
eta = 1
max_iter = 10000

def SOR_object(c_last,N, grid , w, epsilon):
    c_current = c_last.copy()
    c_next = np.zeros((N, N))

    c_current[-1, :] = 0
    c_current[0, :] = 1 
    iteration = 0
    while iteration < max_iter:
        c_next = c_current.copy()  

        for i in range(0, N):
            for j in range(0, N):
                if grid[i, j] == 1: 
                    c_next[i, j] = 0
                elif 0 < i < N-1 and 0 < j < N-1: 
                    c_next[i, j] = (w / 4) * (c_current[i+1, j] + c_next[i-1, j] + c_current[i, j+1] + c_next[i, j-1]) + (1-w) * c_current[i, j]
        #c_next[-1, :] = 0
        #c_next[0, :] = 1
        c_next[:, 0] = c_next[:, 1]
        c_next[:, -1] = c_next[:, -2]

        delta = np.max(np.abs(c_next - c_current))
        if delta < epsilon:
            return c_current, iteration

        c_current[:, :] = c_next[:, :]
        iteration += 1
    return c_current, iteration

def find_candidates(grid):
    candidates = []
    for i in range(N):
        for j in range(N):
            if grid[i, j] == 1: 
                if i + 1 < N and grid[i + 1, j] == 0 and (i + 1, j) not in candidates:
                    candidates.append((i + 1, j))
                if i - 1 >= 0 and grid[i - 1, j] == 0 and (i - 1, j) not in candidates:
                    candidates.append((i - 1, j))
                if j + 1 < N and grid[i, (j + 1) % N] == 0 and (i, j + 1) not in candidates:
                    candidates.append((i, j + 1))
                if j - 1 >= 0 and grid[i, (j - 1)% N] == 0 and (i, j - 1) not in candidates:
                    candidates.append((i, j - 1))
    return candidates

def DLA(candidates,grid,eta):
    #candidates = find_candidates()
    probabilities = np.array([c_current[i, j]**eta for i, j in candidates])
    probabilities /= probabilities.sum() 

    chosen_index = np.random.choice(len(candidates), p=probabilities)
    chosen_point = candidates.pop(chosen_index)
    i, j = chosen_point
    j = j % N  

    grid[i, j] = 1  

    return grid, find_candidates(grid)
# one sinkpoint
```

```python
eta = 1
grid_start  = np.zeros((N,N))
grid_start[-1,50]= 1
#print(grid_start)

c_current = np.zeros((N,N))
c_current[:,] = y.reshape(N, 1)
for i in range(max_growth):
    c_new, iteration = SOR_object(c_current, N, grid_start.copy(), w=1.97, epsilon=1e-5)
    candidates = find_candidates(grid_start)
    #print("Candidates:", candidates)
    grid_new, candidates_new = DLA(candidates,grid_start,eta)
    grid_start = grid_new
grid_end_1_0 = grid_start
c_end_1_ = c_new
```

```python
N = 100
y = np.linspace(N, 1, N)
epsilon = 1e-5
max_iter = 10000
growth_steps = 10
w_values = np.linspace(1.88, 2.0, 6)
avg_iterations_per_w = [] 

for w in w_values:
    grid_start = np.zeros((N, N))
    grid_start[-1, 50] = 1
    c_current = np.zeros((N, N))
    c_current[:,] = y.reshape(N, 1) 

    total_iterations = 0 

    for step in range(growth_steps):
        c_new, iterations = SOR_object(c_current, N, grid_start.copy(), w, epsilon)
        total_iterations += iterations
        candidates = find_candidates(grid_start)
        if not candidates:
            break 
        grid_new, candidates_new = DLA(candidates, grid_start, eta)
        grid_start = grid_new

    avg_iterations = total_iterations / growth_steps 
    avg_iterations_per_w.append(avg_iterations)
    print(f"ω = {w:.2f}, Average Iterations = {avg_iterations:.1f}")

optimal_w = w_values[np.argmin(avg_iterations_per_w)]
print(f"\nOptimal ω for Growth = {optimal_w:.3f}, Min Avg Iterations = {min(avg_iterations_per_w):.1f}")
```

```python
plt.plot(w_values, avg_iterations_per_w, marker='o', linestyle='-')
plt.xlabel("ω value")
plt.ylabel("Avg SOR Iterations Over Growth")
plt.title("Optimizing ω for Growing DLA")
plt.grid()
plt.savefig("Growing_DLA", dpi=300)
plt.show()
```
